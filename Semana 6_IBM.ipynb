```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.dtypes
```

```python
df["peak-rpm"].dtypes
```

```python
#The objective is to find the correlation between the characteristics that are on the complete dataframe.
df.corr()
```

```python
#We can also find the correlation between specific columns, in this case we need to specify the columns and then the formula of correlation.
df[['bore','stroke','compression-ratio','horsepower']].corr()
```

```python
#With this function we have a scatter plot, which shows all the data as a tendency.
sns.regplot(x="engine-size", y="price", data=df)
plt.ylim(0,)
```

```python
#This shows the correlation between engine-size and price.
df[["engine-size", "price"]].corr()
```

```python
#Scattered plot between highway-mpg and price.
sns.regplot(x="highway-mpg", y="price", data=df)
```

```python
df[['highway-mpg', 'price']].corr()
```

```python
sns.regplot(x="peak-rpm", y="price", data=df)
```

```python
df[['peak-rpm','price']].corr()
```

```python
#Also in python we can ask for boxplots, in this case is the distribution according to body-style, showing the distibution among price.
sns.boxplot(x="body-style", y="price", data=df)
```

```python
#And also we can insert a boxplot from a specific column.
sns.boxplot(x="engine-location", y="price", data=df)
```

```python
df['drive-wheels'].value_counts()
```

```python
df['drive-wheels'].value_counts().to_frame()
```

```python
#Here we are grouping the specific columns and then obtained the mean of those grouped values.
df_group_one = df[['drive-wheels','body-style','price']]
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

```python
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
#In order to have a better visualization about data we can convert the complete dataframe on a pivot table.
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

```python
#The color bar graph shows a scalation of the complete data, in order to have a better visualization among all data.
import matplotlib.pyplot as plt

plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
#And also we can obtain the correlation between some columns, showing also the p value.
from scipy import stats
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```
