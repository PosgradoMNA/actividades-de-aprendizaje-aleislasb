```python
#Primero se importa el url donde se encuentra la raw data junto con las librerias correspondientes para ejecutar la limpieza de datos.
#Como resultado tenemos una gran base de datos, los cuales debemos de analizar bajo formulas masivas.
import pandas as pd
import numpy as np
df=pd.read_csv("https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv",index_col=0)
df
```

```python
#Con esta formula estamos preguntando si alguno de los valores encontrados en nuestra tabla de datos se encuentra vacia, ya que lo que queremos es hacer una limpieza al 100% de todos los datos incompletos.
df.isnull().values.any()
```

```python
#Al momento de problarlo con las columnas, vemos que todas las columnas tienen datos faltantes, menos una columna.
df.isnull().any()
```

```python
#Con esta formula estamos indicando que queremos filtrar todos los valores que nos den de nuestra consulta como True, en pocas palabras, que tienen datos faltantes.
df.dropna(inplace=True)
```

```python
#Después de realizar el filtrado, volvemos a realizar la consulta en la cual buscamos si cualquier valor de nuestros datos estan vacios. La respuesta es falsa, esto quiere decir que el filtrado fue el correcto.
df.isna().values.any()
```

```python
#Con esta formula sólo confirmamos que dentro de todas nuestras columnas todos los datos vacíos son filtrados correctamente, desplegando como todos los valores como falsos, ya que todos los resultados tienen al menos un dato.
df.isnull().any()
```
