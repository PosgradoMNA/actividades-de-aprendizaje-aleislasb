```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
```

```python
df.head()
```

```python
#This function replaces the question mark values to NaN.
import numpy as np
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#Evaluates if the dataframe has any missing value. True for missing values and False for non missing values.
missing_data = df.isnull()
missing_data.head(5)
```

```python
#This formula counts the values that have the characteristic of missing or full according to True or False.
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")   
```

```python
#This function converts the column normalized-losses to a number and then results the mean of that column.
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#This formula replaces all the values that contains NaN on the column normalized-losses to the average of that same column.
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#This function does the same calculation of the column bore and obtains the average fo it.
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#Replaces the values NaN to the average.
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)
```

```python
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
df.dtypes
```

```python
#This functions changes the type of data contained on all the rows of the specified columns. 
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
#This confirms that the type of data has been changed.
df.dtypes
```
